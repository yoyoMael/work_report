```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

metric = 'Precision'
def set_values(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width() / 2, height, height, ha='center', va='bottom',fontsize=8)
        rect.set_edgecolor((.9,.9,1))
def plot_graph(metric):
    df = pd.read_csv(metric.lower()+'_comp_dt.csv')

    plt.rcParams['savefig.dpi'] = 300
    plt.rcParams['figure.figsize'] = (12.0, 8.0)
    values = df[df.columns[:]].values
    clf_names = df.columns[:]
    print(values)

    fig = plt.figure()
    width = .35
    index = np.arange(len(clf_names))
    classes = ['FTP-C','FTP-P','P2P','DATABASE']
    for i in range(4):
        ax = plt.subplot(2,2,i+1)
        rects1 = ax.bar(index,values[i],align='center',width=width,
           color=(.4,.6,.6,.5),edgecolor='w',linewidth=1,log=False,label='test 1')
        rects2 = ax.bar(index+width,values[i+4], align='center',width=width,
           color=(.2,.3,.4,1),edgecolor='w',linewidth=1,log=False,label='test 2')
        ax.set_ylabel(metric)
        ax.set_ylim(0,1.05)
        plt.title(classes[i])
        plt.xticks(index+width/2,clf_names)

    # add values on the top of the bar

    fig.legend(['Test1','Test2'])

    plt.savefig('minor_'+metric.lower()+'_comp_dt.png')
    plt.show()
    # from matplotlib.ticker import MultipleLocator, FormatStrFormatter, NullLocator
    # yformatter = FormatStrFormatter('%1.2f')
    # ylocator = NullLocator()
    # ax.yaxis.set_major_formatter(ymajorLocator)
    # ax.yaxis.set_major_locator(ylocator)

    # set_values(rects1)
    # set_values(rects2)
plot_graph('Precision')
plot_graph('Recall')
plot_graph('F1')
```
